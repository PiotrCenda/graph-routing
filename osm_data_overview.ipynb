```python
import osmnx as ox
```

```python
G = ox.graph_from_place("Lisboa", network_type="all", simplify=False)
```

```python
print(type(G))
print(G)
```

```python
G_simplified = ox.simplify_graph(G)
print(G_simplified)
```

```python
fig, ax = ox.plot_graph(G_simplified, node_size=0.5, edge_linewidth=0.1)
```

```python
fig, ax = ox.plot_graph(G_simplified, show=False, save=True, close=True, filepath="./images/lisboa.svg")
```

```python
ox.save_graphml(G_simplified, "./data/lisboa.graphml")
```

```python
ox.settings.all_oneway = True
G_all_oneway = ox.graph_from_place("Lisboa", network_type="all", simplify=True)
ox.save_graph_xml(G_all_oneway, filepath="./data/lisboa.osm")
```
